#### Chart - 6

````python
# Chart - 6 visualization code

# Create Month column
df['Month'] = df['Date'].dt.month_name()

# Order of months
month_order = ['January', 'February', 'March', 'April', 'May', 'June',
               'July', 'August', 'September', 'October', 'November', 'December']

# Plot
plt.figure(figsize=(12,6))

sns.barplot(
    data=df,
    x='Month',
    y='Close',
    order=month_order,
    estimator='mean'
)

plt.title("Average Closing Price by Month")
plt.xlabel("Month")
plt.ylabel("Average Closing Price")
plt.xticks(rotation=45)

plt.show()
````

##### 1. Why did you pick the specific chart?

Insights:

• The average closing price varies across different months.

• March, April, and May have relatively higher average closing prices.

• September and November show comparatively lower average closing prices.

• The monthly differences are moderate, indicating that stock prices fluctuate throughout the year rather than remaining constant.

Yes. These insights can help investors and analysts understand monthly price behavior and identify periods of relatively stronger or weaker performance. This information can support investment planning and market trend analysis.
